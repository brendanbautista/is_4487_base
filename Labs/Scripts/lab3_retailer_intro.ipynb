## Prepare Data

➡️ Assignment Tasks
- Convert the Total_Amount into a number (not a string)
- Remove rows with empty (NULL) values
- Optional: Locate outliers in any of the variables.   Filter them out if they are mistakes; keep them if they are exceptional cases.

```python
df['total_amount'] = df['total_amount'].replace('[\$,]', '', regex=True).astype(float)
df = df.dropna()
df.head()
```

## Summarize Data

➡️ Assignment Tasks
- Display the average price per month
- Display the total revenue for all months

```python
#average price
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

months_order = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
average_price_per_month = df.groupby('order_month')['total_amount'].mean()
average_price_per_month = average_price_per_month.reindex(months_order)

label_encoder = sl.preprocessing.LabelEncoder()
df['order_month_encoded'] = label_encoder.fit_transform(df['order_month'])

average_price_stats = df.groupby('order_month_encoded').agg(
    Average_Price=('product_price', lambda x: pd.to_numeric(x.replace('[\$,]', '', regex=True)).mean())
).reset_index()

average_price_stats['order_month'] = label_encoder.inverse_transform(average_price_stats['order_month_encoded'])

plt.figure(figsize=(8, 6))
plt.bar(average_price_stats['order_month'], average_price_stats['Average_Price'], color='red')
plt.title('Average Price Per Month')
plt.xlabel('Month')
plt.ylabel('Average Price ($)')
plt.xticks(rotation=45)
plt.show()
```

```python
#total revenue
total_revenue_stats = df.groupby('order_month_encoded').agg(
    Total_Revenue=('total_amount', 'sum')
).reset_index()

total_revenue_stats['order_month'] = label_encoder.inverse_transform(total_revenue_stats['order_month_encoded'])

plt.figure(figsize=(8, 6))
plt.bar(total_revenue_stats['order_month'], total_revenue_stats['Total_Revenue'], color='purple')
plt.title('Total Revenue Per Month')
plt.xlabel('Month')
plt.ylabel('Total Revenue ($)')
plt.xticks(rotation=45)
plt.show()
```
